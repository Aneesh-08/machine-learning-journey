## Feature and Target Selection


We separate the dataset into:
- **X (Input Feature):** CGPA
- **y (Target Variable):** Package

We use double brackets for X to keep it 2D, as required by Scikit-learn models.


```python
x = df[['cgpa']]   # Input feature (2D)
y = df['package']  # Target variable (1D)
```

```python
print(x ,'\n')  # Printing our Input Data

print(y ,'\n')  # Printing our output Data
```

```python
# Importing required libraries:

# train_test_split -> to split dataset into training and testing sets
# LinearRegression -> to create and train the linear regression model

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error      # For mean absolute error
from sklearn.metrics import mean_squared_error       # For mean squared error
from sklearn.metrics import root_mean_squared_error  # For Root mean squared error
from sklearn.metrics import r2_score                 # For R2 score
```

## Train-Test Split

To evaluate our model properly, we split the dataset into:
- **Training Set (80%)** → Used to train the model
- **Testing Set (20%)** → Used to evaluate model performance on unseen data

We use `train_test_split` from Scikit-learn.

```python
X_train , X_test, Y_train ,Y_test = train_test_split(x,y,test_size = 0.2 , random_state=30)

print(X_train.shape)

model = LinearRegression()
model.fit(X_train, Y_train) # training the model

predicted_values = model.predict(X_test)  # Predicting the values

mae = mean_absolute_error(Y_test , predicted_values)
print(mae)

mse = mean_squared_error(Y_test , predicted_values)
print(mse)

rmse = root_mean_squared_error(Y_test , predicted_values)
print(rmse)

r2 = r2_score(Y_test , predicted_values)
print(r2)

num = (1-r2) * (len(Y_test) - 1)
dem = len(Y_test) - 1 - (len(df.columns)-1)
adj_r2 = 1 - (num/dem)
print(adj_r2)



```
