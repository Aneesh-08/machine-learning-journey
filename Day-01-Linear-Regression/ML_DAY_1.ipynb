###Importing Libraries

### Load Dataset

```python
df = pd.read_csv('Placement_data.csv')
print(df.head(),'\n')         # Shows the top 5 rows to get a quick look at the data.
```

## 5. Graphical Analysis (Visual EDA)

```python
print ("Below is the Scatter plot between CGPA and Placement-:" , '\n')

plt.scatter(df['cgpa'] , df['package'])     # Ploting a scatter plot between CGPA and Package of student
plt.xlabel("CGPA of Student")
plt.ylabel("Package")
```

## Feature and Target Selection


We separate the dataset into:
- **X (Input Feature):** CGPA
- **y (Target Variable):** Package

We use double brackets for X to keep it 2D, as required by Scikit-learn models.


```python
x = df[['cgpa']]   # Input feature (2D)
y = df['package']  # Target variable (1D)
```

```python
print(x ,'\n')  # Printing our Input Data

print(y ,'\n')  # Printing our output Data
```

```python
# Importing required libraries:
# train_test_split -> to split dataset into training and testing sets
# LinearRegression -> to create and train the linear regression model

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## 🔀 Train-Test Split

To evaluate our model properly, we split the dataset into:
- **Training Set (80%)** → Used to train the model
- **Testing Set (20%)** → Used to evaluate model performance on unseen data

We use `train_test_split` from Scikit-learn.

```python
X_train , X_test, Y_train ,Y_test = train_test_split(x,y,test_size = 0.2 , random_state=30)

print(X_train,'\n')
print(Y_train, '\n')

print("below is the Testing data:",'\n')

print(X_test,'\n')
print(Y_test,'\n')
```

```python
# Creating the Linear Regression model and training (fitting) it on the training data
# The model learns the relationship between CGPA (X_train) and Package (Y_train)

model = LinearRegression()    # this creates the object called model
model.fit(X_train, Y_train)   # this trains the model using training data (80 %)

# --------- MODEL WILL LEARN------------
# Package = b0 + b1 * CGPA
```

```python
# Making predictions using the trained Linear Regression model
# Here, we predict the package for a student with CGPA = 9.8
# model.predict() returns the estimated output based on the learned regression line

ans_1 = model.predict([[9.8]])  # this is for predicting any random value
ans_2 = model.predict(X_test)   # this is to predict for test dataset

print(np.round(ans_1, 2),'\n') # rounding off the values to 2 ddcimal places
print(np.round(ans_2, 2),'\n')
```

```python
# Visualizing the Linear Regression model
# Scatter plot -> shows actual data points (CGPA vs Package)
# Red line -> regression line (best fit learned from training data)
# It helps us see how well the model captures the relationship between CGPA and Package

plt.scatter(df['cgpa'], df['package'], color='blue')    # original data (X vs y)

plt.plot(X_train, model.predict(X_train), color='red')  # regression line using training data

plt.xlabel("CGPA")
plt.ylabel("Package")
plt.title("Regression Line (Best Fit for Original Data)")
plt.show()
```

```python
# Extracting model parameters
# coef_      -> slope (weight) of the input feature (CGPA)
# intercept_ -> bias (constant term) of the regression equation
# Final learned equation: Package = (slope * CGPA) + bias

slope = model.coef_
print(slope , "this is the slope or weight of the input feature",'\n')

bias = model.intercept_
print(bias, "this is the intercept of the model")
```

```python
# Verifying the regression equation mathematically
# Using the formula:  y = (slope * x) + bias
# Here, we manually compute the predicted package for CGPA = 9.8
# This should match the value obtained from model.predict([[9.8]])

k = slope * 9.8 + bias
print(np.round(k, 2))
```
