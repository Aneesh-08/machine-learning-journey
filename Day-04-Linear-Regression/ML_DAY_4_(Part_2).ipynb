## Feature and Target Selection


We separate the dataset into:
- **X (Input Feature):** many different features
- **y (Target Variable):** price

We use double brackets for X to keep it 2D, as required by Scikit-learn models.


```python
X = df.drop(['price', 'flight','source_city'], axis=1)  # feature selection
y = df['price']                                         # Output

X = pd.get_dummies(X, drop_first=True)  # one-hot encoding

print(X.head())
print(X.columns)
```

```python
print(X ,'\n')  # Printing our Input Data

print("This is the price of the filght:"'\n',y,'\n')  # Printing our output Data
```

```python
# Importing required libraries:

# train_test_split -> to split dataset into training and testing sets
# LinearRegression -> to create and train the linear regression model

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score                 # For R2 score
```

## Train-Test Split

To evaluate our model properly, we split the dataset into:
- **Training Set (80%)** → Used to train the model
- **Testing Set (20%)** → Used to evaluate model performance on unseen data

We use `train_test_split` from Scikit-learn.

```python
X_train , X_test, Y_train ,Y_test = train_test_split(X,y,test_size = 0.2 , random_state=30)

print(X_train.shape)

model = LinearRegression()

model.fit(X_train, Y_train) # training the model

predicted_values = model.predict(X_test)  # Predicting the values

r2 = r2_score(Y_test , predicted_values)
print(r2)
```

```python
# finding the value of weights for each input feature

weights = model.coef_
print(weights)

# conept --> for every input feature there will be one weight associated with it

bias = model.intercept_
print("this is the bias of the model")
print(bias)        # this is the intercept value which will be constant and remain one any num of input features
```

##NOW THIS IS MY OWN CLASS

```python
class MyLR:
  def __init__(self):
    self.coef_ = None
    self.intercept_ = None

  def fit(self , X_train , y_train ):

    X_train = np.insert(X_train , 0 , 1  , axis = 1)   #adding a bias term in the matrix

    weights = np.linalg.pinv(X_train) @ y_train

    self.intercept_ = weights[0]
    self.coef_ = weights[1:]

    print("the weights are:")
    print (self.coef_)
    print("the intercept is: ",self.intercept_)

  def predict(self , X_test):
    y_pred = np.dot(X_test ,self.coef_) + self.intercept_
    return y_pred.astype(int)
```

```python
obj = MyLR()

X_train_np = X_train.values.astype(float)
Y_train_np = Y_train.values.astype(float)
X_test_np = X_test.values.astype(float)

obj.fit(X_train_np , Y_train_np)
print("")
y_pred = obj.predict(X_test_np) # this is the price of the flight which is predicted based on multiple input features
r2 = r2_score(Y_test , y_pred)
print(r2)
```
